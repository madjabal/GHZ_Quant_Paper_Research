```python
equal_weight['month'] = equal_weight['date']
```

```python
equal_weight['month'].head()
```

```python
for i in range(len(equal_weight['month'])):
    equal_weight['month'].loc[i] = str(equal_weight['month'].loc[i])[-4:-2]
```

```python
equal_weight['month'].head()
```

```python
for i in range(len(equal_weight['month'])):
    equal_weight['month'].loc[i] = int(equal_weight['month'].loc[i])
```

```python
equal_weight.rename({'date':'year'}, inplace=True, axis=1)
```

```python
equal_weight.head()
```

```python
for i in range(len(equal_weight['year'])):
    equal_weight['year'].loc[i] = int(str(equal_weight['year'].loc[i])[:-4])
```

```python
peqw = pd.read_csv('peqw.csv')
```

```python
corr = peqw.corr()
```

```python
corr.describe()
```

```python
for dtype in ['float','int','object']:
    selected_dtype = peqw.select_dtypes(include=[dtype])
    mean_usage_b = selected_dtype.memory_usage(deep=True).mean()
    mean_usage_mb = mean_usage_b / 1024 ** 2
    print("Average memory usage for {} columns: {:03.2f} MB".format(dtype,mean_usage_mb))
```
